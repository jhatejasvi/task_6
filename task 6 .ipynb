# Sample Collection 

```python
import cv2
cap = cv2.VideoCapture(0)
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
def face_extractor(img):
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray, 1.3, 5)
    
    if faces is ():
        return None
    
    #crop all face found
    for (x,y,w,h) in faces:
        cropped_face = img[y:y+h, x:x+w]
        
    return cropped_face

count = 0

# collect 1000 samples of yr face from webcam
while True:
    
    ret, frame = cap.read()
    if face_extractor(frame) is not None:
        count += 1
        face = cv2.resize(face_extractor(frame), (200,200))
        #face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        
        #save file in specified dir with unique name
        
        file_name_path = './faces/user/1/' + str(count) + '.jpg'
        cv2.imwrite(file_name_path, face)
        
        #put count on imgs & disp live count
        
        cv2.putText(face, str(count), (100,100), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
        cv2.imshow('Face Cropper', face)
        
    else:
        print("Face not found")
        pass
    
    if cv2.waitKey(1) == 13 or count == 10000:
        break
        
cap.release()
cv2.destroyAllWindows()
print("done")
```

# Model Training 

```python
import os
import cv2
import numpy 

def faceDetection(test_img):             
    gray_img=cv2.cvtColor(test_img,cv2.COLOR_BGR2GRAY)
    face_haar=cv2.CascadeClassifier(r'haarcascade_frontalface_default.xml')
    faces=face_haar.detectMultiScale(gray_img,scaleFactor=1.2,minNeighbors=3)
    return faces,gray_img
#Training will begin from here
faces=[]
faceID=[]
    

for path,subdirnames,filenames in os.walk(r'faces\user'):
    for filename in filenames:
        if filename.startswith("."):
            print("skipping system file")
            continue
        id=os.path.basename(path)
        img_path=os.path.join(path,filename)
#         print ("img_path",img_path)
#         print("id: ",id)
        test_img=cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        display_img = test_img
        if test_img is None:
            print ("Not Loaded Properly")
            continue
        cv2.putText(display_img, "training model", (20, 150), cv2.FONT_HERSHEY_COMPLEX, 0.5, (0,0,255), 1)
        cv2.imshow('hello',display_img)
        if cv2.waitKey(1)==13:
            break
        faces.append(test_img)
        faceID.append(int(id))
        model=cv2.face.LBPHFaceRecognizer_create()
cv2.destroyAllWindows()
model.train(faces,numpy.array(faceID))
model.save(r'faces\training_data.yml') 
print("model trained successfully")
```
